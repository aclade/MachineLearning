```python
fhs = pd.read_csv('data_fahasa.csv')
gr = pd.read_csv('good_read.csv')
rd = pd.read_csv('review_data.csv')
```

```python
def no_accent_vietnamese(s):
    s = re.sub('\xa0',' ',s)
    s = re.sub('[áàảãạăắằẳẵặâấầẩẫậ]', 'a', s)
    s = re.sub('[ÁÀẢÃẠĂẮẰẲẴẶÂẤẦẨẪẬ]', 'A', s)
    s = re.sub('[éèẻẽẹêếềểễệ]', 'e', s)
    s = re.sub('[ÉÈẺẼẸÊẾỀỂỄỆ]', 'E', s)
    s = re.sub('[óòỏõọôốồổỗộơớờởỡợ]', 'o', s)
    s = re.sub('[ÓÒỎÕỌÔỐỒỔỖỘƠỚỜỞỠỢ]', 'O', s)
    s = re.sub('[íìỉĩị]', 'i', s)
    s = re.sub('[ÍÌỈĨỊ]', 'I', s)
    s = re.sub('[úùủũụưứừửữự]', 'u', s)
    s = re.sub('[ÚÙỦŨỤƯỨỪỬỮỰ]', 'U', s)
    s = re.sub('[ýỳỷỹỵ]', 'y', s)
    s = re.sub('[ÝỲỶỸỴ]', 'Y', s)
    s = re.sub('đ', 'd', s)
    s = re.sub('Đ', 'D', s)
    return s

def twodtooned(m,kt):
    ma = []
    for lt in m:
        try:
            if kt:
                ma.append(no_accent_vietnamese(lt))
            else:
                ma.append(lt)
        except:
            ma.append('')
    return ma

list_name = twodtooned(fhs['Ten'],True)
temp = twodtooned(rd['name'],True)

        
```

```python
list_inf = []
list_genres = []
def find_index(theList,item):
    for ind in range(len(theList)):
        if item in theList[ind]:
            return ind
    return 'Not Found' 

for i in range(len(temp)):
    res = find_index(list_name,temp[i])
    if res != 'Not Found':
        list_inf.append(fhs['Thong tin'][res])
        list_genres.append(gr['genre'][res])
    else:
        list_inf.append('')
        list_genres.append('')
        
```

```python
dict_pro={'id':rd['id'],'rating':rd['rating'],'name':temp,'Thong tin':list_inf,'The Loai':list_genres}
df=pd.DataFrame(dict_pro)
```

```python
df.to_csv('cainaylacaigi.csv')
```
